## 1. Riders gender throw duration of the trip and age analysis 

### 1.1. Gender with riders age investigation

```python
plt.figure(figsize=[11, 6]);
plt.suptitle('The Age with gender investigation');
plt.subplot(1, 2, 1);
sorted_vals = bike.member_gender.value_counts()/bike.shape[0];
explode = (0.1, 0.1, 0.1)
plt.pie(sorted_vals, labels=sorted_vals.index, explode=explode, startangle=90, counterclock=False, autopct='%1.1f%%');

plt.subplot(1, 2, 2);
base_color = sb.color_palette()[0]
sb.boxplot(data=bike, x='member_gender', y='age', color=base_color);
plt.xlabel('Gender');
plt.ylabel('Member Age');
```

The age of femal riders much narrower than the other two classes, female riders data is much less than male riders and decreases in old age.

### 1.2. Gender with duration in minutes investigation

```python
plt.figure(figsize=[11, 6]);
plt.suptitle('The Duration with gender investigation');

plt.subplot(1, 2, 1);
sorted_vals = bike.member_gender.value_counts()/bike.shape[0];
explode = (0.1, 0.1, 0.1)
plt.pie(sorted_vals, labels=sorted_vals.index, explode=explode, startangle=90, counterclock=False, autopct='%1.1f%%');

plt.subplot(1, 2, 2);
base_color = sb.color_palette()[0]
sb.boxplot(data=bike, x='member_gender', y='duration_minute', color=base_color);
plt.xlabel('Gender');
plt.ylabel('Duration in minutes');
```

The trips for gender is nearly the same duration, but male trips take less duration means they are much quicker.

### 1.3. distribution of the riders gender throw duration of the trip and age 

```python
plt.suptitle('The Age with duration distribution in each gender ');

gender = sb.FacetGrid(data = bike, col = 'member_gender', col_wrap = 3, size = 5,
                 xlim = [19, 72], ylim = [0, 60])
gender.map(plt.scatter, 'age', 'duration_minute', alpha= 0.2)
gender.set_xlabels('Age')
gender.set_ylabels('Duration in minutes')
plt.show()
```

- The number of the riders in concentrated in the middle-left and decresses with increasing the age.
- The duration of the trips in female riders decreases from age 35.
- The number of the female riders decreases as increasing age more than males riders.
- By increasing the age after 40 the number of the rider dereases and the trip duration deareases also.
- Riders from 30 to 40 takes a shorter duration in their trips.
- The other is the least riders and concentated between 22 to 40 with a the shortest duration in their trips.

## 2. The Hourly Usage during Weekdays for Customers and Subscribers

### 2.1. User type with hour of the day investegation

```python
plt.figure(figsize=[11, 6]);
plt.suptitle('The ditribution of user type throw the hours of the day ');

plt.subplot(1, 2, 1);
explode=(0.1,0.1);
sorted_vals = bike.user_type.value_counts()/bike.shape[0];
plt.pie(sorted_vals, labels=sorted_vals.index, explode=explode, startangle=90, counterclock=False, autopct='%1.1f%%');

plt.subplot(1, 2, 2);
day = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
daycat = pd.api.types.CategoricalDtype(ordered=True, categories=day)
bike['start_day'] = bike['start_day'].astype(daycat)
sb.countplot(data=bike, x='start_day', hue='user_type');
plt.xticks(rotation=90)
plt.xlabel('Day of Week');
plt.ylabel('Number of Members');
```

### 2.2. User type with day of the week investigation

```python
plt.figure(figsize=[11, 6]);
plt.suptitle('The ditribution of user type throw which day of the week ');
plt.subplot(1, 2, 1);
sorted_vals = bike.user_type.value_counts()/bike.shape[0];
explode=(0.1, 0.1);
plt.pie(sorted_vals, explode=explode, labels=sorted_vals.index, startangle=90, counterclock=False, autopct='%1.1f%%', shadow=True);

plt.subplot(1, 2, 2);
sb.countplot(data=bike, x='start_hour', hue='user_type');
plt.xticks(rotation=90)
plt.xlabel('Hour of Day');
plt.ylabel('Number of Members');
```

According to the investigation draws about the subscribers that a use the bike sharing system for work commnute thus most trips were on work days (Mon-Fri) and especially during rush hours unlike the customer that nearly the same riding trips throw the hole week but take long trips in weekend.

### 2.3. Heatmap for the hourly usage during weekdays for Customers and Subscribers

```python
plt.figure(figsize=[10, 8]);
plt.suptitle('The hourly usage during weekdays for Customers and Subscribersk ');
# the Hourly Usage during Weekdays for Subscribers
plt.subplot(2, 1, 1)
subscribers = bike.query('user_type == "Subscriber"')
s = subscribers.groupby(['start_day', 'start_hour']).size()
s = s.reset_index(name='count')
s = s.pivot(index='start_day', columns='start_hour', values='count')
sb.heatmap(s, cmap='Blues');
plt.title('Subscribers', loc='left');
plt.xlabel('Hour of Day');
plt.ylabel('Day of Week');
# the Hourly Usage during Weekdays for Customers
plt.subplot(2, 1, 2)
customers = bike.query('user_type == "Customer"')
c = customers.groupby(['start_day', 'start_hour']).size()
c = c.reset_index(name='count')
c = c.pivot(index='start_day', columns='start_hour', values='count')
sb.heatmap(c, cmap='Blues');
plt.title('Customer', loc='left');
plt.xlabel('Hour of Day');
plt.ylabel('Day of Week');
```

- For subscribers: the number of riders concentated in (Mon-to-Fri)to hit the peak in Tuesday and Thursday, and concentated in early hours a day (06:30 to 09:00) and the afternoon (15:30 to 18:30), leaving a conlusion that a subscribers ride for hiting a target or goal or work comitment.
- For customers: Have the same features of the Subscribers in number of the riders oncentated in (Mon-to-Fri)in the same rush houres like subscribers, but for weekends (sat, sun) theirs also a densty in the graph between 9:30 to 17:00.

## 3. The app. distance VS avg. duration for customers and subscribers throw the week

```python
plt.figure(figsize=[11, 5]);
plt.suptitle('The distance VS avg. duration for customers and subscribers throw the weekk ');
# the App. Distance during Weekdays on Subscribers and Customers
plt.subplot(1, 2, 1);
sb.pointplot(data=bike, x='start_day', y='app_distance_km', hue='user_type', linestyles="");
plt.ylim(0, 2.2)
plt.xticks(rotation=90)
plt.xlabel('Day of week');
plt.ylabel('approximated distance in KM');
# the Duration during Weekdays on Subscribers and Customers
plt.subplot(1, 2, 2);
sb.pointplot(data=bike, x='start_day', y='duration_minute', hue='user_type', linestyles="");
plt.ylim(0, 20)
plt.xticks(rotation=90)
plt.xlabel('Day of Week');
plt.ylabel('Avg. Duration in Minute');
```

- The short period of usage for subscribers with concentration on rush hours Mon to Fri, indicating the use is primarily for work commute. The long ride for customer that they're taking advantage of the bike sharing system quite differently from the subscribers, and concentration of customer trips in weekends and in the afternoon.
- The subscriber take a short time in the trips but according to the distance graph means that the subscriber take shorter time and much quecker than customers, and the duration for trips increases in saturday and sunday for the two types.
